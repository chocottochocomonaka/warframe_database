```python
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
res=requests.get("https://steamcharts.com/app/230410#7d")
soup=BeautifulSoup(res.text,"html.parser")
tags=soup.find_all("td",class_="right num-f")
data=[x.string for x in tags]
data = list(map(float, data))#datafloat
data=data[::-1]#
# 
plt.figure(figsize=(12, 6))
plt.plot(data)
```
